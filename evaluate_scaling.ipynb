```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder_path  = "fmif/log/debug/"
file_names = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
file_names = [f for f in file_names if (f[0:len("reward_SVDD_1_LDDT")]=="reward_SVDD_1_LDDT" and f[-3:]=="npz")]

#Load Result from SVDD 
SVDD_LDDT_list_1 =[]
SVDD_LDDT_list_2 =[]
SVDD_LDDT_list_5 =[]
SVDD_LDDT_list_10 =[]
SVDD_LDDT_list_20 =[]
SVDD_LDDT_list_40 =[]
SVDD_LDDT_list_80 =[]

for i in file_names:
    SVDD_LDDT_list_1.append(np.mean(np.load("fmif/log/debug/" + i)['reward']))
    if np.mean(np.load("fmif/log/debug/" + i)['reward'])< 7:
        print(i)
    SVDD_LDDT_list_2.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_2_LDDT" + i[len("reward_SVDD_2_LDDT"): ])['reward']))
    SVDD_LDDT_list_5.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_5_LDDT" + i[len("reward_SVDD_5_LDDT"): ])['reward']))
    SVDD_LDDT_list_10.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_10_LDDT" + i[len("reward_SVDD_5_LDDT"): ])['reward']))
    SVDD_LDDT_list_20.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_20_LDDT" + i[len("reward_SVDD_5_LDDT"): ])['reward']))
    SVDD_LDDT_list_40.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_40_LDDT" + i[len("reward_SVDD_5_LDDT"): ])['reward']))
    SVDD_LDDT_list_80.append(np.mean(np.load("fmif/log/debug/" + "reward_SVDD_80_LDDT" + i[len("reward_SVDD_5_LDDT"): ])['reward']))



```

```python
import seaborn as sns
sns.set_theme()
x = np.log10([59.48,81.28,154.88, 276.39, 514,983.57,1938])
y = np.array([np.mean(SVDD_LDDT_list_1),np.mean(SVDD_LDDT_list_2),np.mean(SVDD_LDDT_list_5),np.mean(SVDD_LDDT_list_10),np.mean(SVDD_LDDT_list_20), np.mean(SVDD_LDDT_list_40),np.mean(SVDD_LDDT_list_80)])
plt.plot(x, y)
plt.scatter(x, y, c="r" )
plt.xlabel('Inference Time (Log 10, s)', fontsize=16)  # X-axis title
plt.ylabel('LDDT', fontsize=16)
plt.savefig("scaling.png")
plt.show()
```
