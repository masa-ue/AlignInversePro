```python
sample_path = '/data/wangc239/proteindpo_data/AlphaFold_model_PDBs/5KPH.pdb'
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Load a molecule
mol = Chem.MolFromPDBFile(sample_path)

# Draw the molecule
Draw.MolToImage(mol)
```

```python
import py3Dmol

# Open the PDB file and read its contents
with open(sample_path, 'r') as file:
    pdb_content = file.read()

# Create a viewer and add the PDB content
view = py3Dmol.view(width=800, height=400)
view.addModel(pdb_content, "pdb")

# Style the molecular visualization
view.setStyle({'cartoon': {'color': 'spectrum'}})

# Zoom to fit and show the viewer
view.zoomTo()
view.show()
```

```python
import py3Dmol

# Read the content of the PDB files
with open(sample_path, 'r') as file:
    pdb1 = file.read()

with open(folded_path, 'r') as file:
    pdb2 = file.read()

# Create a viewer and set its size
viewer = py3Dmol.view(width=800, height=400)

# Add the first PDB
viewer.addModel(pdb1, 'pdb')
viewer.setStyle({'model': 0}, {'cartoon': {'color': 'blue'}})  # Color it blue

# Add the second PDB
viewer.addModel(pdb2, 'pdb')
viewer.setStyle({'model': 1}, {'cartoon': {'color': 'red'}})  # Color it red

# Zoom to fit and center the view
viewer.zoomTo()

# Show the viewer
viewer.show()
```
