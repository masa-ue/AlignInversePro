```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

Load Result from SVDD

```python
folder_path  = "fmif/log"
file_names = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
file_names = [f for f in file_names if (f[0:len("reward_SVDD_scRMSD")]=="reward_SVDD_scRMSD" and f[-3:]=="npz")]

#Load Result from SVDD 
SVDD_LDDT_list =[]
original_LDDT_list =[]
for i in file_names:
    SVDD_LDDT_list.append(np.mean(np.load("fmif/log/" + i)['reward']))
    df = pd.read_csv("fmif/log/reward_original_stability"+ i[len("reward_SVDD_scRMSD"):-3]+"csv")
    original_LDDT_list.append(np.mean(df['bb_rmsd']))
```

```python
plt.scatter(original_LDDT_list,SVDD_LDDT_list)
plt.plot(np.linspace(0,1.0,100),np.linspace(0,1.0,100),c='red')
plt.xlabel("Best of N")
plt.ylabel("SVDD")
plt.show()
```
